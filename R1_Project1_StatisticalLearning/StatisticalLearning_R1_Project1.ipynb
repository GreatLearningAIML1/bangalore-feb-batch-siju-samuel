# 1 Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data. (2.5 points)

```python
## Reading a dataset from a csv file using pandas.
df = pd.read_csv('Laliga.csv', skiprows=1, header=0)
df.tail(5)
```

```python
#set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.
df.replace('-', 0, inplace=True)
df.tail(3)
```

```python
#Check the current datatypes of frame
df.info()
```

```python
#Change to numeric datatype from objects
df = df.apply(pd.to_numeric, errors='ignore')
df.info()
```

# 2 Print all the teams which have started playing between 1930-1980. (5 points)

```python
#some Debue rows are in format 1930-42, take only first year and convert to integer.
debut_parsed = df['Debut'].apply(lambda x:int(x[:4]))
teams = df.loc[(debut_parsed >= 1930) & (debut_parsed <= 1980)]['Team'].tolist()

#Note: Considered 1930 and 1980 also.
print("Teams started playing between 1930-1980 : ", str(teams)[1:-1]) #[1:-1] to remove the brackets in list
print("Total number of teams  : ", len(teams))
```

# 4 Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. (5 points)

Goal_diff_count = GoalsFor - GoalsAgainst

```python
def Goal_diff_count():
    columns = ['Team', 'GoalDiff']
    diff_df = pd.DataFrame(columns=columns)
    for index, row in df.iterrows():
        diff = row.GoalsFor - row.GoalsAgainst
        x = pd.Series([row['Team'], diff], columns)
        diff_df = diff_df.append(x, ignore_index=True)
    diff_df = diff_df.apply(pd.to_numeric, errors='ignore')
    #maxGoalDiff
    val = diff_df.GoalDiff.abs() #abs bcoz max goal diff can happen in -ve values also
    max_goal_diff = val.max()
    t1 = str(diff_df.iloc[[val.idxmax()]]['Team'].tolist())[2:-2]#get team and format
    #minGoalDiff, considering absoulute difference
    val = diff_df.loc[diff_df.GoalDiff.abs() > 0]['GoalDiff'].abs()
    min_goal_diff = val.min()
    t2 = str(diff_df.iloc[[val.idxmin()]]['Team'].tolist())[2:-2]#get team and format
    #minGoals, GoalsFor-GoalsAgainst
    val = diff_df.GoalDiff
    goalsFor_GoalsAgainst = val.min()
    t3 = str(diff_df.iloc[[val.idxmin()]]['Team'].tolist())[2:-2]#get team and format
    print("Maximum Goal Difference: ", max_goal_diff, "Team = ", t1 )
    print("Minimum Goal Difference: ", min_goal_diff, "Team = ", t2)
    print("Goal Difference(GoalsFor - GoalsAgainst): ", goalsFor_GoalsAgainst, "Team = ", t3)

    print()#Blank
    print("Team and Goal Differences  (Note: Negative indicates GoalsAgainst > GoalsFor)")
    print(diff_df.to_string(index=False))
Goal_diff_count()
```

# 5. Create a new column with name “Winning Percent” and append it to the data set (5 points)

Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
df['Winning Percent'] = ((df.GamesWon/df.GamesPlayed) * 100).fillna(value=0)
df.tail(3)
```

```python
print("Top 5 Teams with highest winning percent")
df.nlargest(5, 'Winning Percent')['Team'].tolist()
```
